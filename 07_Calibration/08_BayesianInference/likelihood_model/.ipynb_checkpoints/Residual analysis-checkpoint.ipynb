```python
def getMeasuredPC6Data(dbName):
    cwd = os.getcwd()

    conn = psycopg2.connect("host=localhost dbname=AMS user=postgres password=ia09")
    cur = conn.cursor()
    cur.execute(
        """ SELECT "postcode", "l_pc6_consumption_kwh_m3" FROM """ + str(dbName) + """;
        """
        )
    data = cur.fetchall()
    colnames = [desc[0] for desc in cur.description]
    retrieved = pd.DataFrame(data, columns=colnames)
    cur.close()
    conn.commit()

    return retrieved
```

# One hot encoding

```python
def one_hot_encoding(df_measurement):
    df = pd.DataFrame(columns = ['postcode', 'year', 'EUI'])
    for row in df_measurement.values:
        postcode = row[0]
        measurements = row[1:]
        year = 2009
        for year_record in measurements:
            year += 1
            if not math.isnan(year_record):
                df = df.append({'postcode': postcode, 'year': year,'EUI':year_record}, ignore_index=True)
    return df
```

# Regression

```python
X = df_encoded.drop(['EUI','year_2010'], axis='columns')
y = df_encoded.EUI
```

```python
residuals = []

model = LinearRegression()
model.fit(X,y)
print "model score: ", model.score(X,y)
index = 0
for i in X.values:
    y_i = y[index]
    index += 1
    y_hat = model.predict([i])
    res = y_i-y_hat

    residuals.append(res.tolist()[0])
    
res_array = np.array(residuals)
print 'redistual std: ', np.std(res_array)
```

```python
plt.figure(figsize=(10,7))
sns.distplot(residuals)
plt.xlabel("y_{i} - hat\y_{i} residuals (kWh/m3)", fontsize=14)
plt.title("normalized")
```
